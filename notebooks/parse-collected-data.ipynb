```python
from datetime import datetime as dt
from datetime import datetime, timedelta, timezone
from datetime import datetime


store_list = []
JST = timezone(timedelta(hours=+9), 'JST')

for index,reccord in data.iterrows():
    print(type(reccord['innerdto']['storeList']))
#     for log['storeList'] in  reccord['innerdto']:
    for store in reccord['innerdto']['storeList']:
        store['input_file_name'] = reccord['input_file_name']
        store['date'] = dt.strptime(str(reccord['yyyymmdd']),'%Y%m%d')
        store['day_of_week'] = reccord['day_of_week']
        store['raw_time'] = reccord['input_file_name'].split('_')[2] + reccord['input_file_name'].split('_')[4].split('.')[0]
        store['time_UTC'] = dt.strptime(store['raw_time'], '%Y%m%d%H%M%S').replace(tzinfo=timezone.utc)
        store['datetime_JST'] = store['time_UTC'].astimezone(JST)
        store['time_JST'] = store['time_UTC'].astimezone(JST).time()
        store_list.append(store)
```
